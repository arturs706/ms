```python
import psycopg2
import uuid
import random
from datetime import datetime

# Connect to your PostgreSQL database
conn = psycopg2.connect(
    dbname="user_db",
    user="arturs",
    password="arturs",
    host="0.0.0.0",
    port="5433"
)

# Create a cursor object
cur = conn.cursor()

def generate_user_info():
    user_id = str(uuid.uuid4())
    name = "User " + str(random.randint(1, 100000022))
    username = "username" + str(random.randint(1, 100000022))
    uk_phone = "+44 7" + str(random.randint(1, 9)) + str(random.randint(0, 9)) + "-" + ''.join(str(random.randint(0, 9)) for _ in range(8))
    passwd = "password" + str(random.randint(1, 1000))
    acc_level = random.choice(['trainee', 'junior', 'senior'])
    status = random.choice(['active', 'suspended'])
    return (user_id, name, username, uk_phone, passwd, acc_level, status)


# Insert random data entries into the table
for _ in range(3003):
    entry = generate_user_info()
    cur.execute("""
        INSERT INTO staff_users (user_id, name, username, mob_phone, passwd, acc_level, status) 
        VALUES (%s, %s, %s, %s, %s, %s, %s);
    """, entry)

# Commit changes
conn.commit()

# Close cursor and connection
cur.close()
conn.close()

print("10 random data entries inserted successfully.")
```
